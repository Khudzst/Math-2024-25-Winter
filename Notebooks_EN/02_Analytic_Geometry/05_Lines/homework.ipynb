```python
# The line passes through points A(1,2) and B(3,4). Find the equation of the line.

import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
y_coords = [2, 4]
x_coords = [1, 3]
plt.figure()
plt.plot(1, 2, 'ro', label='A(1, 2)')
plt.plot(3, 4, 'bo', label='B(3, 4)')
plt.plot(x_coords, y_coords, label='Line AB')
plt.grid(True)
plt.xlim=(0,5)
plt.ylim=(0,5)
plt.legend()
plt.show()
```

# the slope formula
m= $\frac{y_2-y_1}{x_2-x_1}$
# two point formula
$y – y1 = m (x – x1)$

```python
# The line passes through point A(1,2) and is parallel to the line y=2x+3. Find the equation of the line.

# the line parallel to y=2x+3 would be 2x
# y= 2x+b, 2=2(1)+b, 2-2=2-2+b, b=0
# y= 2x
```

```python
# Define the equation symbolically
x, y = sp.symbols('x y')
equation_a = sp.Eq(y, 2*x + 3)
equation_b =sp.Eq(y, 2*x)

# Convert to a numerical function
f = sp.lambdify(x, equation_a.rhs)  # rhs extracts the right-hand side
g = sp.lambdify(x, equation_b.rhs)
# Generate x-values
x_values = np.linspace(-5, 5, 100)

# Calculate y-values
y_values = f(x_values)
y_values2 = g(x_values)
# Plot the line
plt.plot(x_values, y_values)
plt.plot(1, 2, 'ro', label='A(1, 2)')
plt.plot(x_values, y_values2)
plt.gca().set_aspect('equal')
plt.grid(True)
plt.show()
```

```python
# The line passes through point A(1,2) and is perpendicular to the line y=2x+3. Find the equation of the line.
# using the previous exercise, we take the negative inverse of the line, which is y=-1/2x
# y = -1/2x + b, 2=-1/2(1)+b, 2 + 1/2 = -1/2 + 1/2 + b, b= 5/2 or 2.5
# y = -1/2x+5/2
x, y = sp.symbols('x y')
equation_c = sp.Eq(y, 2*x + 3)
equation_d =sp.Eq(y, -1/2*x+5/2)

h = sp.lambdify(x, equation_c.rhs)
j = sp.lambdify(x, equation_d.rhs)

x_values = np.linspace(-5, 5, 100)

y_values3 = h(x_values)
y_values4 = j(x_values)

plt.plot(x_values, y_values3)
plt.plot(1, 2, 'ro', label='A(1, 2)')
plt.plot(x_values, y_values4)
plt.gca().set_aspect('equal')
plt.xlim=(-10,15)
plt.ylim=(-10,15)
plt.grid(True)
plt.show()
```
