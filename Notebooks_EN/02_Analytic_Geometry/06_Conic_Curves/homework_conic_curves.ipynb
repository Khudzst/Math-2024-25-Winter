The equation of a circle with center $(h,k)$ and radius $r$ is: $$(x-h)^2+(y-k)^2=r^2$$

```python
# Find the equation of a circle with center at point A(1,2) and radius r=3
import sympy as sp
import numpy as np
# using the formula above, we get: (x-1)^2+(y-2)^2=3^2
# geogebra visualisation:
from IPython.display import HTML
geogebra_link = "https://www.geogebra.org/calculator/ctshyemb"
HTML(f'<iframe src="{geogebra_link}" width="500" height="500"></iframe>')
```

```python
# Find the equation of a parabola intersecting the Ox axis at points x=2, x=4, and passing through point y(3)=1.
# the intersecting points are x=2 and x=4, that means that these are the roots
# so, y=a(x-2)(x-4)
# we have y(3)=1, so x=3, x=1
# 1=a(3-2)(3-4), 1=a(1)(-1), a=-1
# substituting a into y=a(x-2)(x-4):
# y=-(x-2)(x-4)=-(x^2-x4-2x+8)=-(x^2-6x+8),
# y=-x^2+6x-8
# geogebra visualisation
geogebra_link2="https://www.geogebra.org/calculator/xtsxrzwn"
HTML(f'<iframe src="{geogebra_link2}" width="500" height="500"></iframe>')
```

```python
# Find the center of the ellipse with the equation x^2+4y^2−4x−16y+16=0.
# grouping the x and y numbers together, switching 16 to the other side
# x^2-4x+4y^2-16y=-16
# now we transform this current equation into the standart form equation
# x^2-4x+4-4y^2-16y=-16+4, (x-2)^2+4y^2-16y=-12, (x-2)^2+4(y^2-4y+4)=-12+4*4, (x-2)^2+4(y-2)^2=4, (x-2)^2/4+(y-2)^2/1=1

```

The general form of an ellipse centered at $(h,k)$ with semi-major axis $a$ and semi-minor axis $b$ is: $$\frac{(x-h)^2}{a^2}+\frac{(y-k)^2}{b^2}=1$$

### resulting equation:
$$\frac{(x-2)^2}{4}+\frac{(y-2)^2}{1}=1$$
